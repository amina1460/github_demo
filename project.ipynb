### Problem Statement and Importance
The problem focuses on predicting the severity of traffic accidents in Kenya based on various environmental, road, and situational factors. Traffic accidents are a significant public safety concern, leading to loss of life, injuries, and economic costs. Understanding the factors contributing to accident severity can help policymakers, urban planners, and law enforcement agencies implement targeted measures to reduce accidents and improve road safety.

Accurate prediction of accident severity is important for:

Resource Allocation: Ensuring timely emergency response and preparedness.

Policy Development: Informing infrastructure improvements and traffic regulations.

Public Awareness: Educating drivers about high-risk conditions.

### Dataset Overview
The dataset contains detailed information about traffic accidents in Kenya, including environmental conditions, road characteristics, and situational factors. Each row represents a unique accident, with the following features:

Features
Accident_ID: Unique identifier for each accident.

Date and Time: Timestamp of the accident occurrence.

Weather_Condition: Weather during the accident (e.g., Fog, Rain, Clear).

Road_Type: Type of road where the accident occurred (e.g., Urban, Highway).

Visibility: Visibility level at the time of the accident.

Temperature_C: Temperature in degrees Celsius.

Traffic_Density: Density of traffic on the road.

Speed_Limit: Maximum speed allowed on the road segment.

Traffic_Signal: Presence of traffic signals (Yes/No).

Lighting_Condition: Lighting conditions during the accident (e.g., Daylight, Night).

Vehicle_Count: Number of vehicles involved in the accident.

Driver_Age: Age of the driver(s) involved.

Alcohol_Involved: Whether alcohol was involved (Yes/No).

County and Road_Name: Location details of the accident.

Police_Response_Time: Time taken for police to respond to the scene.

Target Variable
Severity: The severity level of the accident categorized as Minor, Moderate, or Severe. This is the variable to be predicted.

The dataset provides a comprehensive view of factors influencing traffic accidents, enabling data-driven insights into their severity under different conditions.

Correlation analysis

```python
import numpy as np
import scipy.stats as stats

# Select only numerical columns for correlation analysis
numerical_cols = ["Visibility", "Temperature_C", "Traffic_Density", "Speed_Limit", 
                  "Vehicle_Count", "Driver_Age", "Police_Response_Time"]

# Compute correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Compute p-values to assess statistical significance
p_values = np.zeros((len(numerical_cols), len(numerical_cols)))

for i in range(len(numerical_cols)):
    for j in range(len(numerical_cols)):
        _, p_values[i, j] = stats.pearsonr(df[numerical_cols[i]], df[numerical_cols[j]])

# Convert p-values to DataFrame for readability
p_values_df = pd.DataFrame(p_values, index=numerical_cols, columns=numerical_cols)

correlation_matrix, p_values_df

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title and display
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

Varriance Inflation Factor analysis

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Prepare the data for VIF calculation (excluding non-numerical columns)
X = df[numerical_cols]

# Compute VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

Chi- Square test

```python
from scipy.stats import chi2_contingency

# Selecting categorical columns
categorical_cols = ["Severity", "Weather_Condition", "Road_Type", "Traffic_Signal", 
                    "Lighting_Condition", "Alcohol_Involved", "County", "Road_Name"]

# Performing chi-square tests between all pairs of categorical variables
chi2_results = []
for i in range(len(categorical_cols)):
    for j in range(i + 1, len(categorical_cols)):
        contingency_table = pd.crosstab(df[categorical_cols[i]], df[categorical_cols[j]])
        chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)
        chi2_results.append((categorical_cols[i], categorical_cols[j], chi2_stat, p_val))

# Convert results into a DataFrame
chi2_df = pd.DataFrame(chi2_results, columns=["Variable 1", "Variable 2", "Chi-Square Statistic", "P-Value"])

# Filter significant results (p-value < 0.05)
significant_results = chi2_df[chi2_df["P-Value"] < 0.05]

chi2_df, significant_results
```

```python
import streamlit as st
import pandas as pd
import pickle

# Load your trained model
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

# Title
st.title(" Accident Severity Predictor")

# User inputs
weather = st.selectbox("Weather Condition", ["Clear", "Rain", "Fog", "Snow"])
traffic_density = st.slider("Traffic Density", 0, 100, 50)
driver_age = st.number_input("Driver Age", min_value=16, max_value=100, value=30)
alcohol_involved = st.radio("Alcohol Involved?", ["No", "Yes"])

# Preprocess inputs (match model input format)
weather_encoded = {"Clear": 0, "Rain": 1, "Fog": 2, "Snow": 3}[weather]
alcohol_binary = 1 if alcohol_involved == "Yes" else 0

input_data = pd.DataFrame([{
    "Weather_Condition": weather_encoded,
    "Traffic_Density": traffic_density,
    "Driver_Age": driver_age,
    "Alcohol_Involved": alcohol_binary
}])

# Predict button
if st.button("Predict Severity"):
    prediction = model.predict(input_data)
    severity = {0: "Low", 1: "Medium", 2: "High"}[prediction[0]]
    st.success(f"Predicted Accident Severity: **{severity}**")
```
